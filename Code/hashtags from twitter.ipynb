```python
from nltk.tokenize import word_tokenize
punctuations = [".",",","!","?",";","’","singapore","singaporean"]
with open('NDR2022.txt', 'r', encoding='UTF-8') as f:
        corp = []
        for line in f:
            if line != '\n':
                line=line.strip('\n').lower() # convert to lower case
                tokens = word_tokenize(line)
                tokens = [x for x in tokens if x not in punctuations] #remove punctuations
                
                stop_set = set(stopwords.words('english'))
                tokens = [x for x in tokens if x not in stop_set] # remove stop words

                #stemmer = PorterStemmer()
                #tokens = [stemmer.stem(x) for x in tokens] #stemming
                corp.append(tokens)
                #for word in tokens:
                    #if word not in punctuations:
                        #corp.append(word)
print(corp)
```

```python
# construct vocabulary - unique words in a corpus
vocab2 = corpora.Dictionary(corp)
```

```python
# construct bag of words, i.e., word vectors (vector space model)
bow2 = [vocab2.doc2bow(doc) for doc in corp] #converts each doc into a vector space model

n_topic = 3
lda = models.LdaModel(bow2, id2word = vocab2, num_topics = n_topic)

# print out each topic, where each topic is represented by top n words
print("Topics")
for i in range(n_topic):
    print(i, ':', lda.print_topic(i, topn = 5))
    
# print out 5 documents to see what topics they are discussing too
for i in range(5):
    print('-' * 20)
    print(corp[i])
    print(lda[bow[i]]) #print the probability of the tweet belonging to each topic.
```

```python
import urllib.parse   

def search_tweets(query):
    '''query: Keywords or hashtags or combinations'''
    
    # encode the the query into a URL-safe format.
    query = urllib.parse.quote(query, safe='')
    
    # we want to exclude duplicate tweets, so use a set
    tweet_text = set()   
    
    # understand the paramenters of the search API by reading documentation.
    # Note that the search API has a higher rate limit, 450 calls / 15 mins per app.
    # Therefore we can set a smaller nap time at 2 seconds
    success, result = twitter_api('search/tweets.json?q=' + query + '&tweet_mode=extended&count=100&lang=en', nap = 2)

    # a dictionary is returned, the list of tweets can be accessed with the key 'statuses'
    for tweet in result['statuses']:
        tweet_text.add(tweet['full_text'])

    return tweet_text
```

```python
# try one query
tweet_text = search_tweets('#377A -rt')
print('collected', len(tweet_text), 'unique tweets')

# try more queries
queries = ['377a', '377A','covid-19','ndr2022','t5']

for query in queries:
    tweet_text = tweet_text.union(search_tweets(query + ' -rt'))
    print('collected', len(tweet_text), 'unique tweets')
```

```python
with open('tweets.txt', 'w', encoding='UTF-8') as out_file:
    for text in tweet_text:
        out_file.write(text.replace('\n', ' ') + '\n')
```

```python
def load_corpus(filename):
    docs = [] #each element in docs [] is a tweet
    # again, set encoding to be able to handle non-English file
    with open(filename, 'r', encoding='UTF-8') as f:
        for line in f:
            docs.append(line.strip('\n').lower())  # convert to lower case
    return docs

docs = load_corpus('tweets.txt')
```

```python
from nltk.tokenize import TweetTokenizer
tt = TweetTokenizer()
docs = [tt.tokenize(doc) for doc in docs] #generates a list of lists, each list contains the tokens of each tweet
print(docs[0]) #tokenized tweet 
```

```python
import re

# remove URL
docs = [[token for token in doc if not token.startswith('http')] for doc in docs]

# remove non-word tokens (eg emojis, numbers) but keep hashtags or user mentions
# first define a function to test if a token is valid
def is_valid_token(token):
    # keep hashtag, user mentions
    if ((token.startswith('#') or token.startswith('@')) and len(token) > 2):
        return True
    # if a token contains at least two consective English letters, we assume it's a word
    if re.match(r'.*[a-z]{2,}.*', token):
        return True
    return False

docs = [[token for token in doc if is_valid_token(token)] for doc in docs]
print(docs[0])
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

# remove stop words
stop_set = set(stopwords.words('english'))
docs = [[token for token in doc if token not in stop_set] for doc in docs]

# stemming
stemmer = PorterStemmer()
# only perform stemming on tokens not starting with # or @
docs = [[token if token.startswith('#') or token.startswith('@') else stemmer.stem(token) for token in doc] for doc in docs]

# remove documents with fewer than three words
docs = [doc for doc in docs if len(doc) >= 3]
```

```python
#Topic Modelling
```

```python
from gensim import corpora, models

# construct vocabulary - unique words in a corpus
vocab = corpora.Dictionary(docs)

# construct bag of words, i.e., word vectors (vector space model)
bow = [vocab.doc2bow(doc) for doc in docs] #converts each doc into a vector space model
```

```python
# build the model
n_topic = 10
lda = models.LdaModel(bow, id2word = vocab, num_topics = n_topic)

# print out each topic, where each topic is represented by top n words
print("Topics")
for i in range(n_topic):
    print(i, ':', lda.print_topic(i, topn = 5))
```

```python
# print out 5 documents to see what topics they are discussing too
for i in range(5):
    print('-' * 20)
    print(docs[i])
    print(lda[bow[i]]) #print the probability of the tweet belonging to each topic. 
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

def print_sentiment(model, doc):
    '''
    model: a sentiment model
    doc: a string representing a document
    '''
    print(doc)
    ps = model.polarity_scores(doc)
    print('-' * 20)
    print('Positive:', '%.1f%%' % (ps['pos'] * 100))
    print('Negative:', '%.1f%%' % (ps['neg'] * 100))
    print('Neutral:', '%.1f%%' % (ps['neu'] * 100))
    print('Compound score (-1 to 1):', ps['compound'])
    print('-' * 20 + '\n')
```

```python
# loading a pre-trained polarity model
model = SentimentIntensityAnalyzer()

# test cases
for tweet in tweet_text:
    print_sentiment(model, tweet)
```
